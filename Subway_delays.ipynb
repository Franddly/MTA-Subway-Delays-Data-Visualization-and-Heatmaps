```python
import pandas as pd

df = pd.read_csv("mta_delays.csv") 
df = df.dropna()
```

```python
print(df.columns.tolist())
```

```python
df['month'] = pd.to_datetime(df['month'])
pivot = df.pivot_table(index='line', columns=df['month'].dt.month, values='delays', aggfunc='sum')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(pivot, cmap='Reds', annot=True, fmt=".0f")
plt.title("MTA Subway Delays by Line and Month 2020-2024")
plt.xlabel("Month")
plt.ylabel("Line")
plt.show()
```

```python
pivot = df.pivot_table(index='line', columns='day_type', values='delays', aggfunc='mean')
sns.heatmap(pivot, cmap='Greens', annot=True, fmt=".1f")
plt.xlabel("Weekday/Weekend")
plt.ylabel("Line")
```

```python
plt.figure(figsize=(14,8))
pivot = df.pivot_table(index='reporting_category', columns=df['month'].dt.month, values='delays', aggfunc='sum')
sns.heatmap(pivot, cmap='YlGnBu', annot=True, fmt=".0f")
```
